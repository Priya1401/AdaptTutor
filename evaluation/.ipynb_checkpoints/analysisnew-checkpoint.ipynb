# AdaptTutor User Study Evaluation
This notebook analyzes data from the AdaptTutor within-subjects user study.
Each participant completed 4 coding problems: 2 under a static tutor and 2 under an adaptive tutor.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
from scipy import stats

# Load datasets
sessions = pd.read_csv('../backend/data_exports/sessions.csv')
telemetry = pd.read_csv('../backend/data_exports/telemetry.csv')
surveys = pd.read_csv('../backend/data_exports/surveys.csv')

sessions['start_time'] = pd.to_datetime(sessions['start_time'])
sessions['end_time'] = pd.to_datetime(sessions['end_time'])
telemetry['timestamp'] = pd.to_datetime(telemetry['timestamp'])
telemetry['event_data'] = telemetry['event_data'].apply(lambda x: json.loads(x) if pd.notnull(x) else {})
surveys['responses'] = surveys['responses'].apply(lambda x: json.loads(x) if pd.notnull(x) else {})

print(f"Loaded {len(sessions)} sessions, {len(telemetry)} telemetry events, {len(surveys)} survey responses.")
```

## Data Preparation
Split per-problem surveys into adaptive vs static conditions using the condition field logged with each survey.

```python
# Extract per-problem surveys
problem_surveys = surveys[surveys['survey_type'].str.startswith('post_problem')].copy()
problem_surveys['problem_id'] = problem_surveys['responses'].apply(lambda x: x.get('problem_id'))
problem_surveys['frustration'] = problem_surveys['responses'].apply(lambda x: int(x.get('frustration', 0)))
problem_surveys['helpfulness'] = problem_surveys['responses'].apply(lambda x: int(x.get('helpfulness', 0)))
problem_surveys['self_reported_state'] = problem_surveys['responses'].apply(lambda x: x.get('self_reported_state', ''))
problem_surveys['condition'] = problem_surveys['responses'].apply(lambda x: x.get('condition', ''))

# Separate by condition
adaptive_surveys = problem_surveys[problem_surveys['condition'] == 'adaptive']
static_surveys = problem_surveys[problem_surveys['condition'] == 'static']

print(f"Per-problem surveys: {len(adaptive_surveys)} adaptive, {len(static_surveys)} static")
```

## H1: Task Completion Rate
Compare the percentage of problems successfully solved under each condition.

```python
# A successful submission exists if we have a post_problem survey for that problem
# (survey only shows on success). Count completions per condition per session.
completions = problem_surveys.groupby(['session_id', 'condition']).size().reset_index(name='completed')
completions['total'] = 2  # 2 problems per condition
completions['completion_rate'] = completions['completed'] / completions['total'] * 100

adaptive_rates = completions[completions['condition'] == 'adaptive']['completion_rate']
static_rates = completions[completions['condition'] == 'static']['completion_rate']

print(f"Adaptive completion rate: {adaptive_rates.mean():.1f}% (SD={adaptive_rates.std():.1f})")
print(f"Static completion rate: {static_rates.mean():.1f}% (SD={static_rates.std():.1f})")
print(f"Difference: {adaptive_rates.mean() - static_rates.mean():.1f}%")
print(f"Target threshold: >= 15% improvement")

fig, ax = plt.subplots(figsize=(6, 4))
means = [static_rates.mean(), adaptive_rates.mean()]
sds = [static_rates.std(), adaptive_rates.std()]
bars = ax.bar(['Static', 'Adaptive'], means, yerr=sds, capsize=5, color=['#6b7280', '#3b82f6'])
ax.set_ylabel('Completion Rate (%)')
ax.set_title('H1: Task Completion Rate by Condition')
ax.set_ylim(0, 110)
for bar, m in zip(bars, means):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 3, f'{m:.1f}%', ha='center', fontsize=11)
plt.tight_layout()
plt.show()
```

## H2: Self-Reported Frustration
Compare mean frustration ratings (1-5 Likert) between conditions using a paired test.

```python
# Compute mean frustration per session per condition
frust_by_session = problem_surveys.groupby(['session_id', 'condition'])['frustration'].mean().reset_index()
frust_pivot = frust_by_session.pivot(index='session_id', columns='condition', values='frustration').dropna()

print(f"Adaptive mean frustration: {frust_pivot['adaptive'].mean():.2f} (SD={frust_pivot['adaptive'].std():.2f})")
print(f"Static mean frustration: {frust_pivot['static'].mean():.2f} (SD={frust_pivot['static'].std():.2f})")
print(f"Difference: {frust_pivot['static'].mean() - frust_pivot['adaptive'].mean():.2f}")
print(f"Target threshold: >= 1.0 point reduction")

# Paired test
if len(frust_pivot) >= 5:
    stat, p = stats.wilcoxon(frust_pivot['static'], frust_pivot['adaptive'])
    print(f"\nWilcoxon signed-rank test: W={stat:.2f}, p={p:.4f}")
else:
    print(f"\nOnly {len(frust_pivot)} paired samples, skipping statistical test (need >= 5).")

fig, ax = plt.subplots(figsize=(6, 4))
means = [frust_pivot['static'].mean(), frust_pivot['adaptive'].mean()]
sds = [frust_pivot['static'].std(), frust_pivot['adaptive'].std()]
bars = ax.bar(['Static', 'Adaptive'], means, yerr=sds, capsize=5, color=['#6b7280', '#3b82f6'])
ax.set_ylabel('Mean Frustration (1-5)')
ax.set_title('H2: Self-Reported Frustration by Condition')
ax.set_ylim(0, 5.5)
for bar, m in zip(bars, means):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.15, f'{m:.2f}', ha='center', fontsize=11)
plt.tight_layout()
plt.show()
```

## H3: Tutor Preference
Percentage of participants who preferred each condition in the post-session survey.

```python
post_surveys = surveys[surveys['survey_type'] == 'post'].copy()
post_surveys['preference'] = post_surveys['responses'].apply(lambda x: x.get('preference', ''))
post_surveys['group'] = post_surveys['responses'].apply(lambda x: x.get('group', 'a'))

# Map preference back to actual condition
# group a: first_half=static, second_half=adaptive
# group b: first_half=adaptive, second_half=static
def map_preference(row):
    pref = row['preference']
    grp = row['group']
    if pref == 'first_half':
        return 'static' if grp == 'a' else 'adaptive'
    elif pref == 'second_half':
        return 'adaptive' if grp == 'a' else 'static'
    return 'no_response'

post_surveys['preferred_condition'] = post_surveys.apply(map_preference, axis=1)
post_surveys = post_surveys[post_surveys['preferred_condition'] != 'no_response']

pref_counts = post_surveys['preferred_condition'].value_counts()
total = len(post_surveys)

adaptive_pref = pref_counts.get('adaptive', 0)
static_pref = pref_counts.get('static', 0)

print(f"Preferred Adaptive: {adaptive_pref}/{total} ({adaptive_pref/total*100:.1f}%)")
print(f"Preferred Static: {static_pref}/{total} ({static_pref/total*100:.1f}%)")
print(f"Target threshold: >= 70% prefer adaptive")

fig, ax = plt.subplots(figsize=(5, 4))
counts = [static_pref, adaptive_pref]
bars = ax.bar(['Static', 'Adaptive'], counts, color=['#6b7280', '#3b82f6'])
ax.set_ylabel('Number of Participants')
ax.set_title('H3: Tutor Preference')
for bar, c in zip(bars, counts):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.2, str(c), ha='center', fontsize=12)
plt.tight_layout()
plt.show()
```

## Perceived Helpfulness by Condition

```python
help_by_session = problem_surveys.groupby(['session_id', 'condition'])['helpfulness'].mean().reset_index()
help_pivot = help_by_session.pivot(index='session_id', columns='condition', values='helpfulness').dropna()

print(f"Adaptive mean helpfulness: {help_pivot['adaptive'].mean():.2f} (SD={help_pivot['adaptive'].std():.2f})")
print(f"Static mean helpfulness: {help_pivot['static'].mean():.2f} (SD={help_pivot['static'].std():.2f})")

if len(help_pivot) >= 5:
    stat, p = stats.wilcoxon(help_pivot['adaptive'], help_pivot['static'])
    print(f"\nWilcoxon signed-rank test: W={stat:.2f}, p={p:.4f}")

fig, ax = plt.subplots(figsize=(6, 4))
means = [help_pivot['static'].mean(), help_pivot['adaptive'].mean()]
sds = [help_pivot['static'].std(), help_pivot['adaptive'].std()]
bars = ax.bar(['Static', 'Adaptive'], means, yerr=sds, capsize=5, color=['#6b7280', '#3b82f6'])
ax.set_ylabel('Mean Helpfulness (1-5)')
ax.set_title('Perceived Helpfulness by Condition')
ax.set_ylim(0, 5.5)
for bar, m in zip(bars, means):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.15, f'{m:.2f}', ha='center', fontsize=11)
plt.tight_layout()
plt.show()
```

## Behavioral Metrics: Runs, Errors, and Help Requests by Condition

```python
# Merge telemetry with session condition info from problem surveys
# Build a mapping of (session_id, problem_id) -> condition from problem surveys
cond_map = problem_surveys[['session_id', 'problem_id', 'condition']].drop_duplicates()

# For telemetry we don't have problem_id per event, so use session-level condition from sessions table
events_df = telemetry.merge(sessions[['session_id', 'condition']], on='session_id', how='left')

metrics = []
for cond in ['static', 'adaptive']:
    subset = events_df[events_df['condition'] == cond]
    runs = len(subset[subset['event_type'] == 'run_click'])
    submits = len(subset[subset['event_type'] == 'submit_click'])
    helps = len(subset[subset['event_type'] == 'help_click'])
    deletes = len(subset[subset['event_type'] == 'delete'])
    errors = len(subset[(subset['event_type'] == 'submission_result') & 
                        (subset['event_data'].apply(lambda x: x.get('status', '') != 'success'))])
    metrics.append({'condition': cond, 'runs': runs, 'submits': submits, 
                    'help_requests': helps, 'deletions': deletes, 'errors': errors})

metrics_df = pd.DataFrame(metrics)
print(metrics_df.to_string(index=False))

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col, title in zip(axes, ['runs', 'help_requests', 'errors'], 
                          ['Code Runs', 'Help Requests', 'Errors']):
    vals = [metrics_df[metrics_df['condition'] == 'static'][col].values[0],
            metrics_df[metrics_df['condition'] == 'adaptive'][col].values[0]]
    ax.bar(['Static', 'Adaptive'], vals, color=['#6b7280', '#3b82f6'])
    ax.set_title(title)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## Error Reduction Rate
Computed as (errors_first_half - errors_second_half) / errors_first_half for each session,
comparing whether error frequency decreases more under adaptive vs static.

```python
# For each session, split telemetry into first half and second half by timestamp
error_reduction = []
for sid in sessions['session_id'].unique():
    sess_events = events_df[(events_df['session_id'] == sid) & 
                            (events_df['event_type'] == 'submission_result')]
    if len(sess_events) < 4:
        continue
    midpoint = sess_events['timestamp'].median()
    first = sess_events[sess_events['timestamp'] <= midpoint]
    second = sess_events[sess_events['timestamp'] > midpoint]
    
    e_first = len(first[first['event_data'].apply(lambda x: x.get('status', '') != 'success')])
    e_second = len(second[second['event_data'].apply(lambda x: x.get('status', '') != 'success')])
    
    if e_first > 0:
        reduction = (e_first - e_second) / e_first
    else:
        reduction = 0.0
    
    cond = sessions[sessions['session_id'] == sid]['condition'].values[0]
    error_reduction.append({'session_id': sid, 'condition': cond, 'error_reduction': reduction})

if error_reduction:
    err_df = pd.DataFrame(error_reduction)
    print(err_df.groupby('condition')['error_reduction'].mean())
else:
    print('Not enough data for error reduction analysis.')
```
